A. BINOMIAL DISTUBUTION
> The number of successes in a sequence of n independent experiments.  
> Examples: yes/no, true/false, present/absent, on/off, heads/tails, **subject will stick around/subject will not stick around**

```python
from scipy.stats import binom
import seaborn as sb
import matplotlib.pyplot as plt
import numpy as np
```

```python
p = 0.0

for x in range(10,20):
    p += sum(np.random.binomial(10,0.7,100000)==x)/100000

print(p)
```

```python
data_binom = binom.rvs(n=10,p=0.7,loc=0,size=11)
ax = sb.distplot(data_binom, kde=True, color='g', hist_kws={'linewidth': 25, 'alpha':1})
ax.set(xlabel='binomial', ylabel='frequency')

plt.title('70% Chance')
```

```python
data_binom = binom.rvs(n=10,p=0.8,loc=0,size=11)
ax = sb.distplot(data_binom, kde=True, color='orange', hist_kws={'linewidth': 25, 'alpha':1})
ax.set(xlabel='binomial', ylabel='frequency')

plt.title('80% Chance')
```

```python
data_binom = binom.rvs(n=10,p=0.9,loc=0,size=11)
ax = sb.distplot(data_binom, kde=True, color='r', hist_kws={'linewidth': 25, 'alpha':1})
ax.set(xlabel='binomial', ylabel='frequency')

plt.title('90% Chance')
```

```python
# pmf = probability of a single point
# cmf = cumulative probability

binom.cdf(n=10, p=.8, k=5)    # probablity that half *or fewer* subjects stick around
1 - binom.cdf(n=10, p=.8, k=5)    # probability that more than half stick around
binom.sf(n=10, p=.8, k=5)    # survivor function
```
